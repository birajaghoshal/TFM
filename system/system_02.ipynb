## Instalación herramientas cliente-servidor

Este apartado no pretende ser una guía exhaustiva de instalación de cada una de las partes que integran el sistema de desarrollo. Se especifican las vías de instalación y versiones utilizadas, tanto para el cliente (portátil) como para el servidor (ordenador con GPU instalada).Este punto es importante debido a las dificultades técnicas que se han encontrado en la instalación en el sistema operativo Windows.

### 1. Python

* Cliente

La versión instalada es 3.6.2, instalada por defecto en Linux Mint 18.2, más tarde se integró al ecosistema Anaconda.

* Servidor

En el caso del servidor, para la instalación se recurrió al ejecutable en la [página oficial](https://www.python.org/downloads) tras diversas pruebas con distintas versiones 3.x, fue la versión 3.6.3 con la que se obtuvo una instalación satisfactoria para trabajar con los drivers de la GPU y Anaconda.

Nota:

Se hará uso de entornos virtuales mediante [virtualenv](https://virtualenv.pypa.io/en/stable/) de esta forma podemos tener mucho mejor organizado nuestro código. Básicamente allá donde creemos en el entorno se instalarán las librerías necesarias, puediendo tener varias versiones de las mismas en varios proyectos. Su uso es muy sencillo, y Anaconda soporta este modo de trabajo.

### 4. cudNN

* Servidor

Un paso importante a tener en cuenta es la instalación de [cudNN v6.0](https://developer.nvidia.com/cudnn) un archivo de unos 100Mb, se trata de una libreria para redes neuronales profundas. ES muy importante instalar la versión adecuada, de lo contrario cuando desde un script python (o notebook jupyter) se realice una llamada a tensorflow arrojará un error al no encontar la libreria (.dll). La instalación en este caso se realizó de manera manual con privilegios de administrador. Copiando los archivos en sus respectivas rutas, tomando como base la instalación de Cuda Toolkit 8.x:

|archivo|ruta|
|-------|----------------|
|cudnn64_6.dll|../CUDA/v8.0/bin|
|cudnn.h|../CUDA/v8.0/include|
|cudnn.lib|../CUDA/v8.0/lib/x84|

Nota: Para completar este punto mediante una consola en modo administrador

se copian uno a uno los archivos anteriores en respectivas rutas mediante el comando cp path_origen.archivo

### 6. Git - Github

Git es una herramienta de versionado de archivos. Se puede mantener una traza con los cambios hechos en uno o varios archivos. Se considera una herramienta muy útil y necesaria en cualquier proyecto como este. Por su parte, Github es un servidor remoto (en la nube) en el que se almacenan estos archivos con sus respectivas versiones.

* Cliente

La [instalación](https://git-scm.com/book/en/v2/Getting-Started-Installing-Git) y configuración en un entorno Linux es inmediata y no genera problemas. La versión instalada es 2.7.4 que se obtiene ejecutando, en una consola en Linux, el comando git --version.

* Servidor

En el caso de Windows se trata de un ejecutable que se encuentra [aquí](https://git-for-windows.github.io/) es un proyecto distinto a [Git](https://git-scm.com/), pero el funcionamiento es el mismo. La versión instalada es 2.15.0windows.1 obtenida ejecutando el mismo comando anterior pero en la consolo git-bash.

Para finalizar, se han generado dos claves ssh (cliente y servidor) para conectarlas al [repositorio remoto](https://github.com/jmartu/TFM) en Github. De este modo el desarrollo de las simulaciones es directo, teniendo tan solo que actualizar los repositorios cuando se tengan que calcular las simulaciones en el lado del servidor (GPU).
