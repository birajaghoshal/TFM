### 1. Python

Para la instalación se recurrió al ejecutable en la [página oficial](https://www.python.org/downloads) tras diversas pruebas con distintas versiones 3.x, fue la versión 3.6.3 con la que se obtuvo una instalación satisfactoria para trabajar con los drivers de la GPU y Anaconda.

Nota:

Se hará uso de entornos virtuales mediante [conda](https://conda.io/docs/user-guide/tasks/manage-environments.html) de esta forma podemos tener mucho mejor organizadas las dependencias de nuestro código. Trabajar con entornos es una metodología muy interesante puesto que, cada uno de ellos es un compartimento estanco con respecto al resto, pudiendo instalar una librería en un entorno específico, sin que el resto de entornos se vean afectados por esa instalación.

### 4. cudNN

Un paso importante a tener en cuenta es la instalación de [cudNN v6.0](https://developer.nvidia.com/cudnn) un archivo de unos 100Mb, es una libreria para realizar cálculos de redes neuronales profundas. Es muy importante instalar la versión adecuada, de lo contrario cuando desde un script Python (o Jupyter) se realice una una llamada a Tensorflow arrojará un error al no encontar la libreria (.dll). La instalación en este caso se realizó de manera manual con privilegios de administrador. Copiando los archivos en sus respectivas rutas, tomando como base la instalación de Cuda Toolkit 8.x:

|archivo|ruta|
|-------|----------------|
|cudnn64_6.dll|../CUDA/v8.0/bin|
|cudnn.h|../CUDA/v8.0/include|
|cudnn.lib|../CUDA/v8.0/lib/x84|

Nota: Para completar este punto mediante una consola en modo administrador

se copian uno a uno los archivos anteriores en respectivas rutas mediante el comando cp path_origen.archivo
path_destino.

### 6. Git - Github

Git es una herramienta de versionado de archivos. Se puede mantener una traza con los cambios hechos en uno o varios archivos. Se considera una herramienta muy útil y necesaria en cualquier proyecto como este. Por su parte, Github es un servidor remoto (en la nube) en el que se almacenan estos archivos con sus respectivas versiones.

En el caso de Windows se trata de un ejecutable que se encuentra [aquí](https://git-for-windows.github.io/) es un proyecto distinto a [Git](https://git-scm.com/), pero el funcionamiento es el mismo. La versión instalada es 2.15.0windows.1 obtenida ejecutando el mismo comando anterior pero en la consolo git-bash.

Para finalizar, se han generado dos claves ssh (cliente y servidor) para conectarlas al [repositorio remoto](https://github.com/jmartu/TFM) en Github. De este modo el desarrollo de las simulaciones es directo, teniendo tan solo que actualizar los repositorios cuando se tengan que calcular las simulaciones en el lado del servidor (GPU).
